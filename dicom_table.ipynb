```python
import numpy as np
import os
import dicom
import pandas as pd
```

```python
# get file list
dicom_path = 'kidney/renosono_sample/'
lstFilesDCM = []    
for dirPath, dirNames, fileList in os.walk(dicom_path):
    for fileName in fileList:
        if ".dcm" in fileName.lower():
            lstFilesDCM.append(os.path.join(dirPath,fileName))
```

```python
# unique header
headers = []
for d in lstFilesDCM:
    x = dicom.read_file(d)
    header = x.dir()
    for h in header:
        if h not in headers:
            headers.append(h)
```

```python
# take all value except PixelData
headers = [h for h in headers if h != 'PixelData'] 
all_data = {}
count = 0
for d in lstFilesDCM:
    x = dicom.read_file(d)
    name = d.split('/')[-1]
    row = []
    row.append(name)
    # print('now processed: ', name, '\n')
    for h in headers:
        try:
            row.append(str(x.data_element(h).value))
        except:
            row.append('not found')
    all_data[count] = row
    count += 1
headers.insert(0, 'name')
```

```python
df = pd.DataFrame.from_dict(all_data,orient='index', dtype = 'str')
df.columns = headers
df.to_csv('test.csv')
```
